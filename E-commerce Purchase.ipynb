** Importing python libraries.**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

**Reading the Ecommerce Purchases csv file and setting it to a DataFrame called ecom**

```python
ecom=pd.read_csv('Ecommerce Purchases')
```

** Checking the head of the DataFrame. **

```python
ecom.head()
```

** How many rows and columns are there? **

```python
ecom.info()
```

```python
sns.set_theme(style='white')
sns.histplot(ecom['Purchase Price'],bins=50)
```

**Opera VS Mozilla**

```python
a=ecom['Browser Info'].apply(lambda x: x.split('/')[0])
sns.countplot(data=ecom,x=a)
```

** Most popular CC Provider**

```python
plt.figure(figsize=(7,7))
sns.countplot(data=ecom,x=ecom['CC Provider'],palette='turbo')
plt.xticks(rotation=90)
```

** 3 Most popular email service provider**

```python
a=ecom['Email'].apply(lambda x: x.split('@')[1])
b=a.apply(lambda x:x.split('.')[0]).value_counts().head(3)
mail=b.index
num=b.values
plt.figure(figsize=(7,5))
sns.lineplot(x=mail,y=num, marker='o')
```

**10 most popular Jobs**

```python
df=ecom['Job'].value_counts().head(10)
a=df.index
b=df.values
plt.figure(figsize=(7,5))
sns.set_theme(style='darkgrid')
sns.lineplot(x=a,y=b,marker='o',color='g')
plt.xticks(rotation=90)
```

** Average Purchase Price**

```python
ecom['Purchase Price'].mean()
```

**Highest and Lowest Purchase Price**

```python
ecom['Purchase Price'].max()
```

```python
ecom['Purchase Price'].min()
```

**AM VS PM Purchases**

```python
sns.set_theme(style='white')
sns.countplot(data=ecom,x='AM or PM',palette='terrain')
```

**email of the person with the following Credit Card Number: 4926535242672853**

```python
ecom[ecom['Credit Card']==4926535242672853]['Email']
```

**How many people have American Express as their Credit Card Provider *and made a purchase above $95 ?**

```python
ecom[(ecom['CC Provider']=='American Express') & (ecom['Purchase Price']>95)].count()
```

** CC Expiry Date**

```python
df=ecom['CC Exp Date'].apply(lambda x: '20'+ x.split('/')[-1]).value_counts()
sns.set_theme(style='whitegrid')
sns.lineplot(data=df,x=df.index,y=df.values,marker='o')
```

```python
sns.heatmap(ecom.corr(),cmap='plasma')
```

**Most commonly used languages**

```python
df=ecom['Language'].value_counts()
a=df.index
b=df.values
sns.lineplot(x=a,y=b,marker='o',color='purple')
```

**5 companies with maximum Purchase Price**

```python
df=ecom.sort_values(by='Purchase Price',ascending=False).head()
sns.lineplot(data=df,x='Company',y='Purchase Price',marker='o',markerfacecolor='r',markersize=10)
plt.xticks(rotation=90)
```
